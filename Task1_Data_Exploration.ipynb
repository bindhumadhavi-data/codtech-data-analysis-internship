```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Reload dataset (important if kernel restarted)
file_path = r"C:\Users\DELL\Downloads\yellow_tripdata_2016-01.csv"
df = pd.read_csv(file_path)

# Convert datetime columns
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'], errors='coerce')
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'], errors='coerce')

# Trip duration in minutes
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Clean unrealistic values
df = df[(df['trip_duration'] > 0) & (df['trip_duration'] < 500)]

# Summary
print(df['trip_duration'].describe())

# Plot
plt.figure(figsize=(8,5))
sns.histplot(df['trip_duration'], bins=50, kde=True)
plt.xlim(0, 100)
plt.title("Trip Duration Distribution (<100 minutes)")
plt.xlabel("Minutes")
plt.ylabel("Number of Trips")
plt.show()
```

```python
# Convert to numeric
df['fare_amount'] = pd.to_numeric(df['fare_amount'], errors='coerce')
df['tip_amount'] = pd.to_numeric(df['tip_amount'], errors='coerce')
df['total_amount'] = pd.to_numeric(df['total_amount'], errors='coerce')

# Summary stats
print("Average Fare:", df['fare_amount'].mean())
print("Average Tip:", df['tip_amount'].mean())
print("Average Total Amount:", df['total_amount'].mean())
print("Total Revenue (Jan 2016):", df['total_amount'].sum())

# Plot
plt.figure(figsize=(8,5))
sns.histplot(df['total_amount'], bins=50, kde=True)
plt.xlim(0, 100)
plt.title("Total Trip Revenue Distribution (<$100)")
plt.xlabel("Revenue ($)")
plt.ylabel("Number of Trips")
plt.show()
```

```python
# Extract pickup hour
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# Count trips
trips_per_hour = df['pickup_hour'].value_counts().sort_index()

print(trips_per_hour)

# Plot
plt.figure(figsize=(10,5))
trips_per_hour.plot(kind='bar', color="green", edgecolor="black")

plt.title("Number of Trips by Hour of Day")
plt.xlabel("Hour of Day (0 = Midnight)")
plt.ylabel("Number of Trips")
plt.xticks(rotation=0)
plt.show()
```

```python
# Extract day of week
df['pickup_day'] = df['tpep_pickup_datetime'].dt.dayofweek

# Count trips by day
trips_per_day = df['pickup_day'].value_counts().sort_index()

print(trips_per_day)

# Plot
plt.figure(figsize=(8,5))
trips_per_day.plot(kind='bar', color="purple", edgecolor="black")

plt.title("Number of Trips by Day of Week")
plt.xlabel("Day of Week (0=Mon ... 6=Sun)")
plt.ylabel("Number of Trips")
plt.xticks(rotation=0)
plt.show()
```

```python
# Select numeric columns
numeric_cols = ['trip_distance', 'fare_amount', 'tip_amount', 'total_amount', 'trip_duration']

# Compute correlations
corr = df[numeric_cols].corr()

# Plot
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
